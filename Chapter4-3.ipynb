#Matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib.style
import numpy as np
matplotlib.style.use('ggplot')

x = [1,2,3]
y = [2,4,9]

plt.plot(x,y)
plt.title('MATLAB-style')
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('OOP-style')
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
ax[0].plot(x,y)
ax[0].set_title('OOP-style')
plt.show()
```

```python
x = [1,2,3]
y1 = [2,4,9]
y2 = [3,2,1]

matplotlib.style.use('classic')
fig, ax = plt.subplots()
ax.plot(x,y1, label='test1')
ax.plot(x,y2, label='test2')
ax.set_title('OOP-style')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.legend(loc='best')
fig.savefig('test1.png')
fig.savefig('test1.svg')
plt.show()
```

```python
fig, ax = plt.subplots()
x = [1,2,3]
y = [10,5,1]
ax.barh(x,y,tick_label=['s','b','a'])
```

```python
np.random.seed(123)
mean = 100
sigma = 15
x = np.random.normal(mean,sigma,1000)

fig, ax = plt.subplots()
n,bins,patches = ax.hist(x,bins=25)
```

```python
np.random.seed(123)
matplotlib.style.use('ggplot')
x0 = np.random.normal(0,10,500)
x1 = np.random.normal(0,15,500)
x2 = np.random.normal(0,20,500)

fig, ax = plt.subplots()
label = ['x0','x1','x2']
ax.boxplot((x0,x1,x2),labels=label)
```
